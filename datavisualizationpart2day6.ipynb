```python
import numpy as np # Library for arithemic and scientific operation
import pandas as pd # Library for data cleaning, data manipulation,data selection,descriptive
import matplotlib.pyplot as plt # Library one for data visualization
import seaborn as sns # library two for data visualization
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
plt.style.use('ggplot')
sns.displot(x=df['total_bill'])
```

```python
sns.regplot(x=df['total_bill'],y =df['tip'])
```

```python
sns.scatterplot(x=df['total_bill'], y=df['size'])
```

```python
sns.jointplot(x=df['total_bill'],y =df['tip'])
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

## count plot : This is use to distribute categorical data in pandas

```python
sns.countplot(x = df['day'])
```

```python
sns.countplot(x =df['day'], hue = df['sex'])
```

```python
sns.barplot(x=df['day'], y= df['total_bill']) #the black line indicates that some days you can have more than the usual 
```

```python
sns.barplot(x=df['day'], y=df['total_bill'], estimator=max)#  thsi is to change the average to i.e max, min,e.t.c
```

```python
sns.barplot(x=df['day'], y=df['total_bill'], estimator=max, palette='copper_r') # This palette is to change the color which has different color setting
```

```python
sns.violinplot(x=df['day'], y=df['total_bill'], estimator=max, palette='copper_r')
```

```python
sns.swarmplot(x=df['day'], y=df['total_bill'],palette='copper_r')
```

```python
sns.stripplot(x=df['day'], y=df['total_bill'],palette='copper_r')
```

```python
sns.boxplot(x =df['total_bill']) # the one over the line is called outlayer
```

```python
sns.boxplot(y =df['total_bill'],x=df['day']) # thsis can be used for comparison numeric AND CATEGORICAL
```

```python
df2 =sns.load_dataset('flights')
df2.head()
```

```python
df_pv = pd.pivot_table(df2,index = 'year',columns='month', values= 'passengers')
df_pv
```

```python
plt.figure(figsize = (12,10))
sns.heatmap(df_pv, annot=True, fmt = '0.4g')
```

```python
plt.figure(figsize = (12,10))
sns.clustermap(df_pv, annot=True, fmt = '0.4g')
```
