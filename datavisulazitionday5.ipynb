```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('RealGDP16062023.csv')
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.rename({'index':'year', 'tyear':'period', 'Period':'Agriculture', 'Agriculture':'CropProduction', 'CropProduction':'Liverstock',
       'Livestock':'Forestry', 'Forestry':'Fishing', 'Fishing':'Industry', 'Industry':'MiningAndQuarrying', 'MiningAndQuarrying':'CrudepetroleumAndNaturalGas',
       'CrudePetroleumAndNaturalGas':'CoalMining', 'CoalMining':'MetalOres', 'MetalOres':'QuarryingAndotherMinerals',
       'QuarryingAndOtherMinerals':'Manufacturing', 'Manufacturing':'OilRefining', 'OilRefining':'Cement', 'Cement':'FoodBeverageAndTobacco',
       'FoodBeverageAndTobacco':'TextileAppareAndFootwear', 'TextileApparelAndFootwear':'WoodAndWoodProducts',
       'WoodAndWoodProducts':'pulpPaperAndPaperProducts', 'PulpPaperAndPaperProducts':'ChemicalAndPharmaceuticalProduct',
       'ChemicalAndPharmaceuticalProducts':'NonMetallicProducts', 'NonMetallicProducts':'PlasticAndRubberProducts',
       'PlasticAndRubberProducts':'ElectricalAndElectronics', 'ElectricalAndElectronics':'BasicMetalIronAndSteel',
       'BasicMetalIronAndSteel':'MotorVechiclesAndAssembly', 'MotorVehiclesAndAssembly':'OtherManufacturing',
       'OtherManufacturing':'ElectricityGasSteamAndAirCon', 'ElectricityGasSteamAndAirCon':'WaterSupplySewageWaste',
       'WaterSupplySewageWaste':'Construction', 'Construction':'Services', 'Services':'Trade', 'Trade':'AccommodationAndFoodServices',
       'AccommodationAndFoodServices':'TransportationAndStorage', 'TransportationAndStorage':'RoadTransport',
       'RoadTransport':'RailTransportAndPipelines', 'RailTransportAndPipelines':'WaterTransport', 'WaterTransport':'AirTransport',
       'AirTransport':'TransportServices', 'TransportServices':'PostAndCourierServices', 'PostAndCourierServices':'InformationAndCommunication',
       'InformationAndCommunication':'Publishing', 'Publishing':'MotionPicturesSoundRecordingAndMusicProduction',
       'MotionPicturesSoundRecordingAndMusicProduction':'Broadcasting', 'Broadcasting':'ArtsEntertainmentAndRecreation',
       'ArtsEntertainmentAndRecreation':'FinanceAndInsurance', 'FinanceAndInsurance':'FinancialInstitutions',
       'FinancialInstitutions':'Insurance', 'Insurance':'RealEstate', 'RealEstate':'ProfessionalScientificAndTechnicalServices',
       'ProfessionalScientificAndTechnicalServices':'AdministrativeAndSupportServicesBusinessServices',
       'AdministrativeAndSupportServicesBusinessServices':'PublicAdministration',
       'PublicAdministration':'Education', 'Education':'HumanHealthAndSocialServices', 'HumanHealthAndSocialServices':'OtherServices',
       'OtherServices':'GDP at 2010 Constant Basic Prices', 'GDP at 2010 Constant Basic Prices':'NetTaxesOnProducts',
       'NetTaxesOnProducts':'NetTaxesOnProducts', 'GDP at 2010 Constant Market Prices':'NAN'},axis = 1,inplace=True)
df.head()
```

## line plot

```python
plt.style.use('ggplot')
plt.plot(df2['year'], df2['Agriculture'], ls = '--',lw=3.5,marker = 'o',markevery = 3,mec='purple',mfc='purple')
plt.xlabel('YEAR')
plt.ylabel('GDP')
plt.title('TREND OF AGRICULTURAL GDP IN BILLIONS OVER THE SPAN OF 41 YEARS ')
plt.show()
```

```python
fig = plt.figure(figsize = (10,10))
axes = fig.add_axes([0.1,0.1,0.7,0.4])
plt.style.use('ggplot')
axes.plot(df2['year'], df2['Agriculture'], ls = '--',lw=3.5,marker = 'o',markevery = 3,mec='purple',mfc='purple')
axes.set_xlabel('YEAR', color = 'navy', fontsize = 25)
axes.set_ylabel('GDP',color = 'navy', fontsize = 25)
axes.set_title('TREND OF AGRICULTURAL GDP IN BILLIONS OVER THE SPAN OF 41 YEARS ',color = 'navy', fontsize = 25)
plt.show()
```

# html color code to choose your color

```python
fig = plt.figure(figsize = (10,10))
axes = fig.add_axes([0.1,0.1,0.7,0.4])
plt.style.use('ggplot')
axes.plot(df2['year'], df2['Agriculture'], ls = '--',lw=3.5,marker = 'o',markevery = 3,mec='purple',mfc='purple',color = 'green',label = 'AGRIC')
axes.plot(df2['year'], df2['CrudepetroleumAndNaturalGas'], ls = '--',lw=3.5,marker = 'o',markevery = 3,mec='purple',mfc='purple',color = 'cyan',label = 'EDUC')
axes.plot(df2['year'], df2['Education'], ls = '--',lw=3.5,marker = 'o',markevery = 3,mec='purple',mfc='purple',color = 'black',label = 'CRUDE')
axes.plot(df2['year'], df2['InformationAndCommunication'], ls = '--',lw=3.5,marker = 'o',markevery = 3,mec='purple',mfc='purple',color = '#87ceeb',label='ICT')
axes.set_xlabel('YEAR', color = 'navy', fontsize = 25)
axes.set_ylabel('GDP',color = 'navy', fontsize = 25)
axes.set_title('TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS ',color = 'navy', fontsize = 25)
plt.legend()
plt.show()
```

```python
fig,axes = plt.subplots(2,2, figsize=(15,10))
axes[0,0].bar(df2['year'], df2['InformationAndCommunication'], color= '#87ceeb')
axes[0,0].set_xlabel('YEAR', color = 'navy', fontsize = 12)
axes[0,0].set_ylabel('GDP',color = 'navy', fontsize = 12)
axes[0,0].set_title('TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS ',color = 'navy', fontsize = 12)

axes[1,0].step(df2['year'], df2['Education'], color= '#071019')
axes[1,0].set_xlabel('YEAR', color = 'navy', fontsize = 12)
axes[1,0].set_ylabel('GDP',color = 'navy', fontsize = 12)
axes[1,0].set_title('TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS ',color = 'navy', fontsize = 12)



axes[0,1].scatter(df2['year'], df2['CrudepetroleumAndNaturalGas'], color= '#0101DF')
axes[0,1].set_xlabel('YEAR', color = 'navy', fontsize = 12)
axes[0,1].set_ylabel('GDP',color = 'navy', fontsize = 12)
axes[0,1].set_title('TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS ',color = 'navy', fontsize = 12)


axes[1, 1].hist(df2['Agriculture'], bins=10, color='#FE2EF7')
axes[1, 1].set_xlabel('YEAR', color='navy', fontsize=12)
axes[1, 1].set_ylabel('GDP', color='navy', fontsize=12)
axes[1, 1].set_title('TREND OF VARIOUS SECTOR GDP IN BILLIONS OVER THE SPAN OF 41 YEARS', color='navy', fontsize=12)

plt.show()
```
